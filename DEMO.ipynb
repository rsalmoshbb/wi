#reemweight Package
 reemweight is a Python package that provides a simple and efficient way to initialize the weights of a neural network sequential model using machine learning algorithms (MLAs).The package compares the performance of these MLAs with other weight initialization strategies such as Xavier and provides the user with the best weight initialization strategy for their neural network.

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from reemweight import LogisticToSequential
# Load breast cancer dataset Or your Binary-based classification dataset 
data = load_breast_cancer()
X = data.data
y = data.target

# Create instance of LogisticToSequential class
log_seq = LogisticToSequential(X, y)

# Train logistic regression model
log_seq.train_logistic_regression()

# Create sequential model
log_seq.create_sequential_model(output_units=1)
# Evaluate sequential model
cm = log_seq.evaluate_sequential_model()

# Plot performance
log_seq.plot_performance()
```

```python
log_seq.create_sequential_modelXavier(output_units=1)
# Evaluate sequential model
cm = log_seq.evaluate_sequential_model()

# Plot performance
log_seq.plot_performance()
```
