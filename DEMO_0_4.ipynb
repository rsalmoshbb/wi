#2- Load your dataset as X,y

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
import numpy as np

# Generate the classification dataset
X, y = make_classification(n_samples=5000, n_features=30, n_informative=10,n_redundant=5, random_state=1, class_sep=3.0)

# Choose a random sample of 2000 points
idx = np.random.choice(X.shape[0], size=2000, replace=False)
X_sample = X[idx]
y_sample = y[idx]

# Define the colors for the scatter plot
colors = ['b' if label == 0 else 'orange' for label in y_sample]

# Plot the scatter plot with the specified colors
plt.scatter(X_sample[:, 0], X_sample[:, 1], c=colors)

# Add a legend box to the plot
legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label='Class 0', markerfacecolor='b', markersize=8),
                   plt.Line2D([0], [0], marker='o', color='w', label='Class 1', markerfacecolor='orange', markersize=8)]
plt.legend(handles=legend_elements, loc='upper right')

# Save the plot as a PNG image with a DPI of 300

#Show the plot
plt.show()

```

#3-usage of the reemweight library to select and train an appropriate model based on the characteristics of the input dataset.

1- Import necessary classes from the reemweight library: DataExaminer, MyModel_Linear, and MyModel_NonLinear.

2- Instantiate the DataExaminer class with the input dataset X and the target labels y.

3- Check if the dataset is linearly separable using the check_linear_separability() method of the DataExaminer class.

4- Based on the result from step 3, select the appropriate model for the dataset. If the dataset is linearly separable, choose MyModel_Linear; otherwise, select MyModel_NonLinear.

5- Load the input data and target labels into the selected model using the load_data() method.

6- Train the selected model with different weight initialization strategies: default, hybrid, Xavier, and He initialization. The training process is performed using the train(), train_Hybrid(), train_Xavier(), and train_He() methods.

7- After training the model with each weight initialization strategy, plot the loss and accuracy curves using the PlotLoss() and PlotAccuracy() methods.


```python
from reemweight import DataExaminer, MyModel_Linear, MyModel_NonLinear


# Check if the dataset is linearly separable
examiner = DataExaminer(X, y)
is_linear = examiner.check_linear_separability()

# Select the appropriate model
if is_linear:
   if __name__ == "__main__":
        my_model = MyModel_Linear()#MyModel_Linear()
        my_model.load_data(X,y)
        my_model.train() #train()
        my_model.train_Hybrid()
        my_model.train_Xavier()
        my_model.train_He()
        my_model.PlotLoss()
        my_model.PlotAccuracy()
else:
    if __name__ == "__main__":
        my_model = MyModel_NonLinear()#MyModel_NonLinear()
        my_model.load_data(X,y)
        my_model.train() #train()
        my_model.train_Hybrid()
        my_model.train_Xavier()
        my_model.train_He()
        my_model.PlotLoss()
        my_model.PlotAccuracy()
```
